# Problem 1

```python
text_file = open('./day1.txt','r')
```

```python
ans = []
for line in text_file:
    line_strip = line.strip()
    l, r = 0, len(line_strip) - 1
    l_digit, r_digit = 0, 0
    lfound, rfound = False, False
    
    while l <= len(line_strip) - 1 and r >= 0:
        if line_strip[l].isdigit() and not lfound:
            lfound = True
            l_digit = line[l]
            
        if line_strip[r].isdigit() and  not rfound:
            rfound = True
            r_digit = line[r]
            
        if lfound and rfound:
            combined_num = int(str(l_digit) + str(r_digit))
            ans.append(combined_num)
            break
        
        l+=1
        r-=1

print(ans)
print(sum(ans))
```
